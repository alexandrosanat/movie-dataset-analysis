### Get info box from a single Paramount movie wiki page

For all Paramount movies between 2010 - 2019 we will get the inforbox from their wiki pages

```python
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
import json
```

Load webpage

```python
r = requests.get("https://en.wikipedia.org/wiki/Shutter_Island_(film)")

# Convert to beautiful soup object
soup = bs(r.content)

# Print out the HTML
contents = soup.prettify()
print(contents)
```

After inspecting html page we need to get "infobox vevent" table object

```python
info_box = soup.find(class_="infobox vevent")
print(info_box.prettify())
```

```python
info_rows = info_box.find_all("tr")  # Get all table rowns

for row in info_rows:
    print(row.prettify())
```

```python
movie_info = {}


def get_content_value(row_data):
    if row_data.find("li"):
        return [li.get_text(" ", strip=True).replace("\xa0", " ") for li in row_data.find_all("li")]
    else:
        return row_data.get_text(" ", strip=True).replace("\xa0", " ")


for index, row in enumerate(info_rows):
    if index == 0:
        movie_info["title"] = row.find("th").get_text(" ", strip=True)
    elif index == 1:
        continue
    else:
        content_key = row.find("th").get_text(" ", strip=True)
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value
```

### Get info box from a all Paramount movie wiki pages between 2010 and 2019

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_Paramount_Pictures_films_(2010%E2%80%932019)")

# Convert to beautiful soup object
soup = bs(r.content)

# Print out the HTML
contents = soup.prettify()
print(contents)
```

After inspecting html page we need to get "infobox vevent" table object

```python
movie_summaries = soup.find("table",{"class":"wikitable sortable"})
print(movie_summaries.prettify())
```

```python
movie_info_rows = movie_summaries.find_all("tr")  # Get all table rowns
movie_summaries_dict = {}

for i, row in enumerate(movie_info_rows):
    if (i+1) % 2 == 0:
        # print(row.prettify())
        title = row.find("i").get_text(" ", strip=True)
        url = row.a["href"]
        movie_summaries_dict[title] = url
```

### Wrap up progress into function

```python
def get_info_box(url):

    def get_content_value(row_data):
        if row_data.find("li"):
            return [li.get_text(" ", strip=True).replace("\xa0", " ") for li in row_data.find_all("li")]
        else:
            return row_data.get_text(" ", strip=True).replace("\xa0", " ")
    
    r = requests.get(url)

    # Convert to beautiful soup object
    soup = bs(r.content)

    # Get info box soup
    info_box = soup.find(class_="infobox vevent")

    for index, row in enumerate(info_rows):
        if index == 0:
            movie_info["title"] = row.find("th").get_text(" ", strip=True)
        elif index == 1:
            continue
        else:
            content_key = row.find("th").get_text(" ", strip=True)
            content_value = get_content_value(row.find("td"))
            movie_info[content_key] = content_value
            
    return movie_info
```

```python
all_movie_data = {}
base_path = "https://en.wikipedia.org"

# For each movie in the decade get the info box
for title, url in movie_summaries_dict.items():
    all_movie_data[title] = get_info_box(base_path + url)
```

Save data to a json

```python
def save_json(title, data):
    with open(title, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
```

```python
def load_json(title):
    with open(title, "r") as f:
        return json.load(f)
```

```python
save_json("paramount_movies.json", all_movie_data)
```
